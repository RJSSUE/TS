```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
data = pd.read_csv("total.csv")
data
```

```python
tracks = data.groupby("track_title",as_index=False)["lyric"].apply(lambda x: "\n\n".join(x))
songs = pd.merge(tracks,data[["album_name","track_title","track_n"]],how="inner",on="track_title")
songs.drop_duplicates(inplace=True)
songs
```

```python
song_info = data[["album_name","track_title","track_n"]]
song_info.drop_duplicates(inplace=True)
song_info.to_csv("info.csv")
```

```python
song_info = pd.read_csv("info.csv")
```

```python
for index,row in song_info.iterrows():
    song_name = str(row["track_title"])
    print(song_name)
    file_name = row["release_date"]+"-"+song_name.replace(" ", "-")+".md"
    file = open(file_name,"w")
    header = "---\ntitle: "+song_name+"\ncategories:\n- "+row["album_name"]+"\nfeature_image: "+row["feature_image"]+"\n--- \n"
    contents = tracks[tracks["track_title"]==song_name]["lyric"].values[0]
    video = ""
    if not pd.isna(row["video_id"]):
        print(row["video_id"])
        video = "\n### Video\n\n{% include video.html id=\""+row["video_id"]+"\" %}\n\n"
    file.write(header+contents+video)
    file.close()
```
